# A standard false colour combination

```python
file_name = 'Laptop.tiff'

ds = gdal.Open(file_name)
ds
```

```python
multi_data = ds.ReadAsArray(band_list = [3,2,1])

multi_data.shape
```

```python
multi_data = np.transpose(multi_data, axes = (1,2,0))
```

False color composite
